```python
# importar libreria
import pandas as pd
titanic = pd.read_excel('titanic.xlsx', index_col = 0 )
titanic
```

Descripción del las columnas del DataFrame

* **Id_Pasajero:** Id de cada pasajero.

* **Survived:** esta característica tiene valor 0 y 1. 0 para no sobrevivido y 1 para sobrevivido.

* **Clase:** Hay 3 clases: Clase 1, Clase 2 y Clase 3.

* **Nombre:** Nombre del pasajero.

* **Sexo:** Género del pasajero.

* **Edad:** Edad del pasajero.

* **Familiares:** numero de familiares del pasajero.

* **Ticket:** Número de ticket del pasajero.

* **Tarifa:** Indicando la tarifa.

* **Cabina:** La cabina del pasajero.

* **Embarcado:** La categoría embarcada.

```python
#Promedio de edad 
titanic.groupby(['Sexo','Clase'])[['Edad']].mean()
```

```python
#contar la cantidad de sobreviviente por categoria y sexo
titanic.groupby(['Clase','Sexo'])[['Survived']].count()
titanic.groupby(['Clase','Sexo'])[['Survived']].sum()
```

```python
def tarifa_a_usd_actual(tarifa_libras):
    """
    Convierte tarifa del Titanic en libras de 1912
    a precio de dolar hoy
    """
    if pd.isna(tarifa_libras):
        return None
    # 1 libra 1912 ≈ 125 USD actuales
    if tarifa_libras > 0 :
        return tarifa_libras * 125

titanic["Tarifa_USD_2025"] = titanic["Tarifa"].apply(tarifa_a_usd_actual)
titanic.groupby("Clase")["Tarifa_USD_2025"].agg(["mean","min","max"])

```

```python
#Tarifa máxima pagada por los pasajeros dentro de cada puerto de embarque (Embarcado)
titanic['Tarifa_max_puerto'] = titanic.groupby("Embarcado")["Tarifa"].transform("max")
titanic
```

```python
#promedio de la Tarifa, cruzando las variables Clase (filas) y Sexo (columnas)
tabla_dinamica = titanic.pivot_table(
    index="Clase",
    columns="Sexo",
    values="Tarifa",
    aggfunc={"mean","sum"}
)
tabla_dinamica
```

```python
#Embarcado: El puerto de embarque del pasajero. C = Cherburgo, Q = Queenstown, S = Southampton.
tabulacion_cruzada = pd.crosstab(titanic['Embarcado'], titanic['Clase'])
tabulacion_cruzada
```
